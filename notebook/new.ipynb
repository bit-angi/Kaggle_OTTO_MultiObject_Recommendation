```python
# 1. 比较基本统计信息
print("=== 他人验证集 ===")
print(f"测试集大小: {len(train_df)}")
print(f"标签集大小: {len(train_labels_df)}")
print(f"测试集中的session数: {train_df['session'].nunique()}")
print(f"标签集中的session数: {train_labels_df['session'].nunique()}")
print("\n按type统计标签集：")
print(train_labels_df['type'].value_counts().sort_index())

print("\n=== 我的验证集 ===")
print(f"测试集大小: {len(df2)}")
print(f"标签集大小: {len(label2s_df2)}")
print(f"测试集中的session数: {df2['session'].nunique()}")
print(f"标签集中的session数: {label2s_df2['session'].nunique()}")
print("\n按type统计标签集：")
print(label2s_df2['type'].value_counts().sort_index())
```

```python
# 2. 比较session长度分布
def analyze_session_length(df, name):
    session_lengths = df.groupby('session').size()
    print(f"\n=== {name}的session长度分析 ===")
    print("统计描述：")
    print(session_lengths.describe())
    print("\n分位数分布：")
    print(session_lengths.quantile([0.1, 0.25, 0.5, 0.75, 0.9]))

print("测试集的session长度分析：")
analyze_session_length(train_df, "他人验证集")
analyze_session_length(df2, "我的验证集")
```

```python
# 3. 比较时间分布
print("\n=== 时间范围比较 ===")
print("他人验证集时间范围：")
print(f"测试集：{pd.to_datetime(train_df['ts'].min(), unit='s')} 到 {pd.to_datetime(train_df['ts'].max(), unit='s')}")
# print(f"标签集：{pd.to_datetime(train_labels_df['ts'].min(), unit='s')} 到 {pd.to_datetime(train_labels_df['ts'].max(), unit='s')}")

print("\n我的验证集时间范围：")
print(f"测试集：{pd.to_datetime(df2['ts'].min(), unit='s')} 到 {pd.to_datetime(df2['ts'].max(), unit='s')}")
# print(f"标签集：{pd.to_datetime(label2s_df2['ts'].min(), unit='s')} 到 {pd.to_datetime(label2s_df2['ts'].max(), unit='s')}")
```

```python
# 4. 分析验证集中的行为分布
def analyze_behavior_distribution(test_df, label_df, name):
    print(f"\n=== {name}的行为分布分析 ===")
    
    # 计算每个session的各类型行为数量
    behavior_stats = label_df.groupby('session')['type'].value_counts().unstack(fill_value=0)
    behavior_stats.columns = ['clicks', 'carts', 'orders']
    
    print("\n每个session的行为统计：")
    print(behavior_stats.describe())
    
    # 计算行为转化率
    print("\n行为转化率：")
    total_sessions = len(behavior_stats)
    print(f"有点击的session比例: {(behavior_stats['clicks'] > 0).mean():.3f}")
    print(f"有加购的session比例: {(behavior_stats['carts'] > 0).mean():.3f}")
    print(f"有下单的session比例: {(behavior_stats['orders'] > 0).mean():.3f}")
    
    # 计算平均行为数
    print("\n平均每个session的行为数：")
    print(behavior_stats.mean())

print("比较两个验证集的行为分布：")
analyze_behavior_distribution(train_df, train_labels_df, "他人验证集")
analyze_behavior_distribution(df2, label2s_df2, "我的验证集")
```

```python
# 5. 分析行为的时序性
def analyze_behavior_sequence(test_df, label_df, name):
    print(f"\n=== {name}的行为时序分析 ===")
    
    # 合并测试集和标签集
    merged = pd.concat([
        test_df.assign(set_type='test'),
        label_df.assign(set_type='label')
    ]).sort_values(['session', 'ts'])
    
    # 检查每个session中test和label的时序
    session_analysis = []
    for session_id, group in merged.groupby('session'):
        test_max_ts = group[group.set_type == 'test']['ts'].max()
        label_min_ts = group[group.set_type == 'label']['ts'].min()
        session_analysis.append({
            'session': session_id,
            'is_chronological': test_max_ts < label_min_ts if not pd.isna(test_max_ts) and not pd.isna(label_min_ts) else False,
            'time_gap': label_min_ts - test_max_ts if not pd.isna(test_max_ts) and not pd.isna(label_min_ts) else None
        })
    
    session_df = pd.DataFrame(session_analysis)
    print(f"时序正确的session比例: {session_df['is_chronological'].mean():.3f}")
    print("\n时间间隔统计（秒）：")
    print(pd.Series(session_df['time_gap']).describe())
```

```python
# 执行时序分析
print("分析两个验证集的时序性：")
analyze_behavior_sequence(train_df, train_labels_df, "他人验证集")
analyze_behavior_sequence(df2, label2s_df2, "我的验证集")
```
